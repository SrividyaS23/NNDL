```
from sympy import *
x=symbols('x')
f=x**4+3*x**2+10
z=diff(f,x) 
print(z)
X=int(input("enter x value:"))
itr=1 
n=0.1 
epoch=2
while (itr <= epoch) :
   Y=z.subs(x,X)
   print("z",Y)
   deltax=-n*Y
   print("delta(x) =",deltax)
   newx = X + deltax
   print("x =",newx)
   itr= itr + 1
   X=newx
print("final x =",X)
print("Minimum value is",f.subs(x,X))
```
